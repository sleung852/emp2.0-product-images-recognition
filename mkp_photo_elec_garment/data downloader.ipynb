```python
import pandas as pd
from tqdm import tqdm
import os
from urllib.request import urlretrieve
```

```python
gar=pd.read_excel('gar.xlsx')
elec=pd.read_excel('elec.xlsx')
```

```python
dataset=elec.append(gar)
```

```python
dataset.drop_duplicates('PRODUCT_ID',inplace=True)
```

```python
dataset['URL']=dataset['PRODUCT_ID'].apply(
lambda x: 'http://i03.hktdc-img.com/rsc?profile=productimage&subprofile=enlarge&pid={}&file=image_1.jpg'.format(x))
```

```python
dataset=dataset[['URL','PRODUCT_ID']]
dataset.head()
```

```python
print('Initiating downloading images...')
for _, row in tqdm(dataset.iterrows(), total=dataset.shape[0]):
    urlretrieve(row["URL"], os.path.join('data','{}.jpg'.format(row['PRODUCT_ID'])))
print('Image downloading completed!')
```
